Preliminares:
Desarrollar un modelo predictivo sobre la probabilidad que un individuo presente salarios por sobre o por bajo los 50.000
dólares anuales, en base a una serie de atributos socioeconómicos.
Considerando que lo que queremos saber es si se gana sobre o bajo los 50.000 dólares, estaríamos hablando de una Clasificación. Podríamos decir que ganar sobre los 50.000 dólares anuales es 1 y por debajo de los 50.000 dólares anuales es 0. De esta manera estaríamos hablando de una primera aproximación mediante la binarización de nuestro problema a estudiar.

Las variables explicativas a utilizar en el modelo son las siguientes:

● age: Edad del individuo ● workclass: Naturaleza de la organización que emplea al individuo ● education: Nivel educacional del individuo ● capital-gains: Ingresos generados por inversiones fuera del trabajo asalariado ● capital-losses: Pérdidas generadas por inversiones fuera del trabajo asalariado ● fnlwgt: Ponderador muestral ● marital-status: Estado civil del individuo ● occupation: Ocupación del individuo ● relationship: Relación respecto a su familia ● race: Raza del encuestado ● sex: Sexo del encuestado ● hours-per-week: Cantidad de horas trabajadas por semana ● native-country: País de origen

Dado que la variable objetivo es de tipo categórica se aplicará un modelo predictivo de clasificación (aprendizaje supervisado)

Con respecto a la evaluación del desempeño del modelo de clasificación a aplicar, se utilizará la siguiente métrica:

● Curva de ROC: Permite evaluar el rango de errores del modelo y evalúa la relación entre falsos positivos y verdaderos positivos. - En el eje X va la tasa de Falsos Positivos (falsas alarmas) - En el eje Y va la tasa de Verdaderos Positivos
